```python
import json

class Sage:
    """Wise Healer character to help"""

    def __init__(self, name, variety, lifes, healing=100, attributes_file="dragon_quest_attributes.json", weapons_file="dragon_quest_weapons.json"):
        """Initialize the name, variety, lifes, healer attributes, and load additional attributes and weapons."""
        self.name = name
        self.variety = variety
        self.lifes = lifes
        self.healing = healing
        self.level = 1
        self.xp = 0
        self.load_attributes(attributes_file)
        self.load_weapons(weapons_file)

    def load_attributes(self, file_path):
        with open(file_path, "r") as file:
            data = json.load(file)
            sage_data = data["characters"]["Sage"]
            self.hp = sage_data["HP"]
            self.mp = sage_data["MP"]
            self.attack = sage_data["Attack"]
            self.defense = sage_data["Defense"]
            self.speed = sage_data["Speed"]
            self.magic = sage_data["Magic"]
            self.skills = sage_data["Skills"]
            self.equipment = sage_data["Equipment"]

    def load_weapons(self, file_path):
        with open(file_path, "r") as file:
            self.weapons = json.load(file)["weapons"]

    def perform_attack(self, attack_type):
        attack_types = {
            "Arcane Blast": self.magic + 40,
            "Mystic Fire": self.magic + 35,
            "Spirit Surge": self.magic + 25
        }
        if attack_type in attack_types:
            return f"Sage performs {attack_type}, dealing {attack_types[attack_type]} damage!"
        return "Invalid attack type."

    def defend(self):
        self.defense += 25
        return f"Sage conjures a magical barrier, increasing defense by 25!"

    def heal(self, amount=None):
        if amount:
            self.lifes += amount
            return f"Sage {self.name}, a {self.variety} healer with {self.lifes} life points and {self.healing} healing power."
        if self.mp >= 15:
            self.hp += 30
            self.mp -= 15
            return f"Sage casts Heal, restoring 30 HP!"
        return "Not enough MP to cast Heal."

    def gain_xp(self, amount):
        self.xp += amount
        if self.xp >= self.level * 100:
            self.level_up()

    def level_up(self):
        self.level += 1
        self.hp += 12
        self.mp += 8
        self.attack += 3
        self.defense += 3
        self.magic += 5
        self.xp = 0
        return f"Sage has leveled up to level {self.level}!"

    def display_stats(self):
        return {
            "Level": self.level,
            "XP": self.xp,
            "HP": self.hp,
            "MP": self.mp,
            "Attack": self.attack,
            "Defense": self.defense,
            "Speed": self.speed,
            "Magic": self.magic,
            "Skills": self.skills,
            "Equipment": self.equipment
        }

# Example usage
sage = Sage(name="Eagus", variety="Mystical", lifes=150)
print(sage)
print(sage.heal(50))
print(sage.perform_attack("Arcane Blast"))
print(sage.defend())
print(sage.display_stats())
```

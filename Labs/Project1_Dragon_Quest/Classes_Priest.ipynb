```python
import json
import random

class Character:
    def __init__(self, name, role, attributes_file="Attributes.json", weapons_file="Weapons.json"):
        self.name = name
        self.role = role
        self.level = 1
        self.xp = 0
        self.inventory = []
        self.equipped_weapon = None
        self.weapons_data = {}
        self.load_attributes(attributes_file)
        self.load_weapons(weapons_file)

    def load_attributes(self, file_path):
        try:
            with open(file_path, "r") as file:
                data = json.load(file)
                char_data = data.get("characters", {}).get(self.role, {})
                if not char_data:
                    print(f"Warning: {self.role} not found in {file_path}. Using default attributes.")
                self.hp = char_data.get("HP", 100)
                self.mp = char_data.get("MP", 50)
                self.attack = char_data.get("Attack", 10)
                self.defense = char_data.get("Defense", 5)
                self.speed = char_data.get("Speed", 7)
                self.magic = char_data.get("Magic", 20)
                self.skills = char_data.get("Skills", [])
                self.equipment = char_data.get("Equipment", [])
        except (FileNotFoundError, json.JSONDecodeError):
            print(f"Error loading {file_path}. Using default attributes.")

    def load_weapons(self, file_path):
        try:
            with open(file_path, "r") as file:
                data = json.load(file)
                self.weapons_data = data.get("weapons", {})
                if not self.weapons_data:
                    print("No weapons found in file.")
        except (FileNotFoundError, json.JSONDecodeError):
            print(f"Error loading {file_path}. Using empty weapon list.")
            self.weapons_data = {}

    def find_weapon(self):
        if self.weapons_data:
            found_weapon = random.choice(list(self.weapons_data.keys()))
            if found_weapon not in self.inventory:
                self.inventory.append(found_weapon)
                return f"{self.name} found a {found_weapon}!"
            return f"{self.name} found a {found_weapon}, but already has it."
        return "No weapons available to find."
    
    def equip_weapon(self, weapon_name):
        if weapon_name in self.inventory and weapon_name in self.weapons_data:
            self.equipped_weapon = weapon_name
            weapon_stats = self.weapons_data[weapon_name]
            self.attack += weapon_stats.get("Attack", 0)
            self.defense += weapon_stats.get("Defense Boost", 0)
            self.magic += weapon_stats.get("Magic Boost", 0)
            return f"{self.name} equipped {weapon_name}!"
        return "Weapon not found in inventory or invalid weapon."
    
    def show_inventory(self):
        print(f"{self.name}'s Inventory: {', '.join(self.inventory) if self.inventory else 'Empty'}")
    
    def show_equipped_weapon(self):
        print(f"{self.name} has equipped: {self.equipped_weapon if self.equipped_weapon else 'No weapon equipped'}")

    def perform_attack(self, attack_type):
        if attack_type in self.skills:
            damage = self.magic + random.randint(10, 30)
            return f"{self.name} performs {attack_type}, dealing {damage} damage!"
        return "Invalid attack type."

    def defend(self):
        print(f"{self.name} raises defense!")
        self.defense += 10

    def heal(self):
        if self.mp >= 10:
            self.hp += 20
            self.mp -= 10
            print(f"{self.name} heals 20 HP!")
        else:
            print("Not enough MP to heal.")

    def gain_xp(self, amount):
        self.xp += amount
        if self.xp >= self.level * 100:
            self.level_up()

    def level_up(self):
        self.level += 1
        self.hp += 10
        self.mp += 5
        self.attack += 2
        self.defense += 2
        self.magic += 3
        self.xp = 0
        print(f"{self.name} has leveled up to level {self.level}!")

    def display_stats(self):
        print({
            "Name": self.name,
            "Role": self.role,
            "Level": self.level,
            "XP": self.xp,
            "HP": self.hp,
            "MP": self.mp,
            "Attack": self.attack,
            "Defense": self.defense,
            "Speed": self.speed,
            "Magic": self.magic,
            "Skills": self.skills,
            "Equipment": self.equipment,
            "Inventory": self.inventory,
            "Equipped Weapon": self.equipped_weapon
        })

class Battle:
    def __init__(self, player1, player2):
        self.player1 = player1
        self.player2 = player2

    def start_battle(self):
        print("Battle begins!")
        turn = 0
        while self.player1.hp > 0 and self.player2.hp > 0:
            current_player = self.player1 if turn % 2 == 0 else self.player2
            opponent = self.player2 if turn % 2 == 0 else self.player1
            print(f"{current_player.name}'s turn!")
            action = input("Choose action: (attack/defend/heal): ")
            if action == "attack":
                attack_type = input(f"Choose attack type ({', '.join(current_player.skills)}): ")
                print(current_player.perform_attack(attack_type))
            elif action == "defend":
                current_player.defend()
            elif action == "heal":
                current_player.heal()
            else:
                print("Invalid action. Skipping turn.")
            turn += 1

        winner = self.player1 if self.player1.hp > 0 else self.player2
        print(f"{winner.name} wins the battle!")


def setup_game():
    print("Welcome to Dragon Quest Battle!")
    player1_name = input("Enter Player 1 name: ")
    player1_role = input("Choose Player 1 role: ")
    player1 = Character(player1_name, player1_role)

    player2_name = input("Enter Player 2 name: ")
    player2_role = input("Choose Player 2 role: ")
    player2 = Character(player2_name, player2_role)

    battle = Battle(player1, player2)
    battle.start_battle()

setup_game()

import json
import random       

class Character:
    def __init__(self, name, role, attributes_file="Attributes.json", weapons_file="Weapons.json"):
        self.name = name
        self.role = role
        self.level = 1
        self.xp = 0
        self.inventory = []
        self.equipped_weapon = None
        self.weapons_data = {}
        self.load_attributes(attributes_file)
        self.load_weapons(weapons_file)







    def load_attributes(self, file_path):






    
            try:
                with open(file_path, "r") as file:
                    data = json.load(file)
                    char_data = data.get("characters", {}).get(self.role, {})
                    if not char_data:
                        print(f"Warning: {self.role} not found in {file_path}. Using default attributes.")
                    self.hp = char_data.get("HP", 100)
                    self.mp = char_data.get("MP", 50)
                    self.attack = char_data.get("Attack", 10)
                    self.defense = char_data.get("Defense", 5)
                    self.speed = char_data.get("Speed", 7)
                    self.magic = char_data.get("Magic", 20)
                    self.skills = char_data.get("Skills", [])
                    self.equipment = char_data.get("Equipment", [])
            except (FileNotFoundError, json.JSONDecodeError):
                print(f"Error loading {file_path}. Using default attributes.")
```

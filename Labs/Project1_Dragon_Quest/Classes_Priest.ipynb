```python
class Priest:
    def __init__(self, attributes_file="dragon_quest_attributes.json", weapons_file="dragon_quest_weapons.json"):
        self.level = 1
        self.xp = 0
        self.load_attributes(attributes_file)
        self.load_weapons(weapons_file)
    
    def load_attributes(self, file_path):
        with open(file_path, "r") as file:
            data = json.load(file)
            priest_data = data["characters"]["Priest"]
            self.hp = priest_data["HP"]
            self.mp = priest_data["MP"]
            self.attack = priest_data["Attack"]
            self.defense = priest_data["Defense"]
            self.speed = priest_data["Speed"]
            self.magic = priest_data["Magic"]
            self.skills = priest_data["Skills"]
            self.equipment = priest_data["Equipment"]
    
    def load_weapons(self, file_path):
        with open(file_path, "r") as file:
            self.weapons = json.load(file)["weapons"]
    
    def cast_spell(self, spell):
        if spell in self.skills and self.mp >= 10:
            self.mp -= 10
            return f"Priest casts {spell}!"
        return "Not enough MP or spell not known."
    
    def equip_item(self, item):
        if item in self.weapons and item not in self.equipment:
            self.equipment.append(item)
            return f"Equipped {item}."
        return f"{item} is either already equipped or not a valid item."
    
    def perform_attack(self, attack_type):
        attack_types = {
            "Holy Strike": self.attack + 20,
            "Divine Light": self.magic + 30,
            "Smite": self.attack + 10
        }
        if attack_type in attack_types:
            return f"Priest performs {attack_type}, dealing {attack_types[attack_type]} damage!"
        return "Invalid attack type."
    
    def defend(self):
        self.defense += 20
        return f"Priest enters a defensive stance, increasing defense by 20!"
    
    def divine_guidance(self):
        return "Priest invokes Divine Guidance, revealing hidden paths and granting wisdom!"
    
    def gain_xp(self, amount):
        self.xp += amount
        if self.xp >= self.level * 100:
            self.level_up()
    
    def level_up(self):
        self.level += 1
        self.hp += 10
        self.mp += 5
        self.attack += 2
        self.defense += 2
        self.magic += 3
        self.xp = 0
        return f"Priest has leveled up to level {self.level}!"
    
    def display_stats(self):
        return {
            "Level": self.level,
            "XP": self.xp,
            "HP": self.hp,
            "MP": self.mp,
            "Attack": self.attack,
            "Defense": self.defense,
            "Speed": self.speed,
            "Magic": self.magic,
            "Skills": self.skills,
            "Equipment": self.equipment
        }
```

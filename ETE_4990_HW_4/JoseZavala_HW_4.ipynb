```python


def decode(message_file_path):
    """
    Decodes a message from a pyramid-structured .text file.
    
    :param message_file_path: Path to the input text file
    :return: Decoded message string
    """
    number_word_map = {}
    with open(message_file_path, 'r') as file:
        for line in file:
            if line.strip():
                number, word = line.strip().split(maxsplit=1)
                number_word_map[int(number)] = word

    sorted_numbers = sorted(number_word_map.keys())

    pyramid = []
    index = 0
    row_length = 1

    while index + row_length <= len(sorted_numbers):
        row = sorted_numbers[index:index + row_length]
        pyramid.append(row)
        index += row_length
        row_length += 1

    decoded_words = [number_word_map[row[-1]] for row in pyramid]
    return ' '.join(decoded_words)



if __name__ == "__main__":
    
    message_file = "Hidden_text.text"  

    try:
        decoded = decode(message_file)
        print("Decoded message:")
        print(decoded)
    except FileNotFoundError:
        print(f"[Error] File not found: {message_file}")
    except Exception as e:
        print(f"[Error] Something went wrong: {e}")
```

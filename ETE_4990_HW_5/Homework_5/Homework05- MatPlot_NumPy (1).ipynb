Using the following file https://www.kaggle.com/datasets/gregorut/videogamesales

Create and sort the data by Name, Year, and NA_Sales

````python
import numpy as np
import csv

def load_vgsales(file):
    # Initialize lists to store the data
    names = []
    years = []
    genres = []
    publishers = []
    na_sales = []
    eu_sales = []
    jp_sales = []
    other_sales = []
    global_sales = []
    critic_scores = []
    critic_counts = []
    user_scores = []
    user_counts = []
    developers = []
    ratings = []

    # Open and read the CSV file
    with open(file, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)  # Using DictReader to access columns by name
        
        for row in reader:
            names.append(row['Name'])
            years.append(row['Year_of_Release'])
            genres.append(row['Genre'])
            publishers.append(row['Publisher'])
            na_sales.append(float(row['NA_Sales']) if row['NA_Sales'] else 0)
            eu_sales.append(float(row['EU_Sales']) if row['EU_Sales'] else 0)
            jp_sales.append(float(row['JP_Sales']) if row['JP_Sales'] else 0)
            other_sales.append(float(row['Other_Sales']) if row['Other_Sales'] else 0)
            global_sales.append(float(row['Global_Sales']) if row['Global_Sales'] else 0)
            critic_scores.append(float(row['Critic_Score']) if row['Critic_Score'] else 0)
            critic_counts.append(int(row['Critic_Count']) if row['Critic_Count'] else 0)
            user_scores.append(float(row['User_Score']) if row['User_Score'] else 0)
            user_counts.append(int(row['User_Count']) if row['User_Count'] else 0)
            developers.append(row['Developer'])
            ratings.append(row['Rating'])

    # Convert lists into numpy arrays
    data = np.array(list(zip(names, years, genres, publishers, na_sales, eu_sales, jp_sales, 
                             other_sales, global_sales, critic_scores, critic_counts, 
                             user_scores, user_counts, developers, ratings)), dtype=object)

    return data

# Example usage
file_path = 'vgsales-2.csv'
vgsales_data = load_vgsales(file_path)

# Print a sample of the loaded data
print(vgsales_data[:5])  # Print first 5 rows
````

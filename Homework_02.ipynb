````python
class Animal:
    def __init__(self, name, benefits):
        self.name = name
        self.benefits = benefits
    
    def describe(self):
        return f"{self.name.capitalize()} are beneficial because {self.benefits}."

class Pig(Animal):
    def __init__(self):
        super().__init__("pigs", "they provide meat, skin, and help improve soil biodiversity")

class Cow(Animal):
    def __init__(self):
        super().__init__("cows", "they provide milk, meat, and can be used for labor")

class LLama(Animal):
    def __init__(self):
        super().__init__("llama", "they guard the sheep and assist in carry loads, fur is use for coats")

class Goat(Animal):
    def __init__(self):
        super().__init__("goats", "they provide milk, meat, and help clear overgrown vegetation")

class Chicken(Animal):
    def __init__(self):
        super().__init__("chickens", "they lay eggs, provide meat, and help control insects")
class Duck(Animal):
    def __init__(self):
        super().__init__("ducks", "they serve as pets for the farm and swim on the pond.")

# store animals lsit
animals = [Pig(), Cow(), LLama(), Goat(), Chicken(),Duck()]

def display_animals():
    if not animals:
        print("No animals on the farm.")
    else:
        for animal in animals:
            print(animal.describe())

def add_animal(name, benefits):
    animals.append(Animal(name, benefits))
    print(f"{name.capitalize()} has been added.")

def remove_animal(name):
    global animals
    name = name.lower()
    initial_count = len(animals)
    animals = [animal for animal in animals if animal.name.lower() != name]
    if len(animals) < initial_count:
        print(f"{name.capitalize()} has been removed.")
    else:
        print(f"{name.capitalize()} is not found on the farm.")

while True:
    action = input("Would you like to add, remove, or view animals? (add/remove/view/exit): ").strip().lower()
    if action == "add":
        name = input("Enter the animal's name: ").strip().lower()
        benefits = input(f"Enter the benefits of {name}: ").strip()
        add_animal(name, benefits)
    elif action == "remove":
        name = input("Enter the animal's name to remove: ").strip().lower()
        remove_animal(name)
    elif action == "view":
        display_animals()
    elif action == "exit":
        break
    else:
        print("Invalid input, please try again.")

    # used Chatgpt to help add and remove animals form the list. 
````
